```python
import numpy as np
import pandas as pd
from sklearn import datasets

Boston=datasets.load_boston()
# dataframe
BostonDf=pd.DataFrame(Boston.data, columns=Boston.feature_names)
BostonDf['targer']=Boston.target 
BostonDf.head()
```

```python
import seaborn as sbn
import matplotlib.pyplot as mplt
corr_mx=BostonDf.corr().round(2)
mplt.figure(figsize=(14,12))
sbn.heatmap(data=corr_mx, annot= True, center=0)
```

```python
mplt.scatter(BostonDf['LSTAT'],BostonDf['targer'], color='gray')
mplt.xlabel('LSTAT')
mplt.ylabel('target=Price')
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(BostonDf[['LSTAT']], BostonDf['targer'], random_state=0)

from sklearn.linear_model import LinearRegression
LRmodel= LinearRegression()
LRmodel.fit(X_train,y_train)
Ypred=LRmodel.predict(X_test)

mplt.scatter(X_test, y_test, color='gray')
mplt.plot(X_test, Ypred, color='purple', linewidth=2)
mplt.xticks(())
mplt.yticks(())
mplt.xlabel('LSTAT')
mplt.ylabel('target=Price')
mplt.show()




```

```python
#evaluate the performance of the model using MSE(Mean Squared Error)
from sklearn.metrics import mean_squared_error, r2_score  
print("MSE: {:.2f}".format(mean_squared_error(y_test, Ypred)))  
```
